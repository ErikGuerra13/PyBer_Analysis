```python
#Create a line chart using the MATLAB method
%matplotlib inline
```

```python
#Import dependecies
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to a list of strings for each month. 
x_axis = ["Jan", "Feb", "Mar", "April","May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Creating a line chart using the object-oriented interface

#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Use plt.shpw()

#Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Annotate using MATLAB method

#Create the plot and add a label for the legend. 
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')

#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Set the y limit between 0 and 45
plt.ylim(0, 45)

#Create a title
plt.title("PyBer Fare By Month")

#Add a legend
plt.legend()
```

```python
#Create a vertical bar chart using the MATLAB method

#Create the plot
plt.bar(x_axis, y_axis, color="green", label="Boston")

#Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title
plt.title("PyBer Fare by Month")

#Add the legend
plt.legend()
```

```python
#Create a horizontal bar char

#Create the plot 
plt.barh(y_axis, x_axis, color="magenta", label="Boston")
#invert the y axis
plt.gca().invert_yaxis()

#Create labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare($)")

#Create a title
plt.title("Pyber Fare by Month")

#Add a legend
plt.legend()
```

```python
#Create a bar chart using Object-oriented Approach

#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```
